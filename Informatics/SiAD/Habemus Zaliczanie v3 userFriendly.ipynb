**_Relacja między kształtem funkcji zmiennej opisującej a wartościami współczynników korelacji:_**
- **Jeśli dane są wynikiem równomiernego spróbkowania funkcji o punkcie przegięcia w środku przedziału próbkowania, wartość współczynnika korelacji Kendalla jest co do wartości bezwzględnej nie mniejsza od wartości współczynnika korelacji Pearsona**
- **Jeśli dane są wynikiem równomiernego spróbkowania funkcji o niezerowych wartościach pierwszej pochodnej i równych zeru wartościach drugiej pochodnej, wartość współczynnika korelacji Spearmana jest równa wartości współczynnika korelacji Pearsona**
- **Jeśli dane są wynikiem równomiernego spróbkowania funkcji o niezerowych wartościach pierwszej pochodnej i równych zeru wartościach drugiej pochodnej, wartość współczynnika korelacji Kendalla jest równa wartości współczynnika korelacji Spearmana**

## Fałsz:
Dla testu niesparowanego dla dużej próby n1>=30,n2>=30 wariancje dla testu Z wyliczamy jako sumę kwadratów wariancji dla średnich w poszczególnych próbach

Test niesparowany dla małej próby n1<30,n2<30 wykonujemy z użyciem rozkładu t-studenta o n1+n2-1 stopniach swobody

Testując równość wariancji (w teście F) stosujemy rozkład Fishera-Snedecora o (n1-1)-(n2-1) stopniach swobody, gdzie n1 i n2 oznaczają liczności prób względem statystyki zdefiniowanej jako iloraz wariancji

By móc użyć test sparowany wariancje w obu próbach muszą być identyczne

Znając wyłącznie współczynnik korelacji możemy określić zna wyrazu wolnego b

Jeśli wszystkie obserwacje w próbie leżą na prostej to współczynnik korelacji wynosi 0

Jeśli jedną ze zmiennych przemnożymy przez 2, współczynnik korelacji się zmieni

Znając wyłącznie współczynnik korelacji możemy wyznaczyć współczynnik kierunkowy a

Jeśli zmienną y przemnożymy przez 2, współczynnik kierunkowy a nie zmieni się

Współczynnik korelacji równy zero oznacza, że zmienne są niezależne

Jeśli jedną ze zmiennych przemnożymy przez 2, współczynnik korelacji zmaleje dwukrotnie

Nie udało się odrzucić hipotezy zerowej w teście na istotność współczynnika korelacji. Oznacz to, że zmienne są nieskorelowane

Współczynnik korelacji między x i y nie jest taki sam, jak współczynnik korelacji między y i x 

Współczynnik korelacji między x i y może być różny od współczynnika korelacji między y i x

**W trakcie obliczania korelacji Spearmana:**  
- oblicza się współczynnik Kendalla i przekształca się go na znormalizowaną wartość korelacji  
- przypisuje się rangi różnicom między zmiennymi dla par obserwacji  
- oblicza się przyrosty wartości zmiennych dla kolejnych obserwacji

**Korelacja rangowa Spearmana:**  
- może być zamiennie stosowana z współczynnikiem korelacji Pearsona  
- stosowana w analizie szeregów czasowych do wykrycia autokorelacji zmiennych  

**Test U-Manna Whitneya:**  
- wymaga obliczenia różnic dla par powiązanych obserwacji zmiennych porządkowych  
- pozwala na zbadanie różnicy między średnimi w dwóch porównywanych populacjach  
- ma wzór statystyki U zawierający sumę rang R dla obserwacji z obu porównywanych prób  

**Test Wilcoxona:**  
- jest innym sformułowaniem testu z statystyką o rozkładzie zmiennej U  
- jest stosowany do porównania median w dwóch próbach niezależnych  
- weryfikuje istotność statystyczną wyliczonej wartości korelacji Spearmana

**Wskaż prawdziwe zdania o związku współczynnika korelacji Kendalla z regresją liniową:**  
- jeśli resztowa suma kwadratów odchyleń SSE>0, współczynnik korelacji Kendalla pomiędzy jedyną  zmienną opisującą a zmienną opisywaną jest mniejszy co do wartości bezwzględnej od analogicznego  współczynnika korelacji Pearsona  

**Wskaż prawdziwe zdania o związku współczynnika korelacji Kendalla z regresją liniową:**  
- jeśli pierwotne wartości zmiennej opisywanej są równe ich predykcjom uzyskanym z użyciem modelu regresji liniowej, współczynnik korelacji Spearmana pomiędzy zmienną opisującą a zmienną opisywaną jest równy analogicznemu współczynnikowi korelacji Kendalla  
- jeśli pierwotne wartości zmiennej opisywanej są równe ich predykcjom uzyskanym z użyciem modelu regresji liniowej, współczynnik korelacji Spearmana pomiędzy zmienną opisującą a zmienną opisywaną jest co do wartości bezwzględnej większy od analogicznego współczynnikowi korelacji Pearsona  
- w przypadku stwierdzenia heteroskedastyczności róźnic między pierwotnymi wartościami zmiennej a ich predykcjami uzyskanymi z użyciem modelu regresji liniowej, nie należy stosować współczynnika korelacji Spearmana  

**Wskaż prawdziwe zdania o stosowalności poszczególnych typów współczynników korelacji:**
- w przypadku stwierdzenia heteroskedastyczności różnic między pierwotnymi wartościami zmiennej opisywanej a ich predykcjami uzyskanymi z użyciem modelu regresji liniowej, nie należy stosować współczynnika korelacji Kendalla
- w przypadku stwierdzenia niezgodności rozkładu różnic między pierwotnymi wartościami zmiennej opisywanej a ich predykcjami uzyskanymi z użyciem modelu regresji liniowej, nie należy stosować współczynnika korelacji Spearmana

**_Relacja między kształtem funkcji zmiennej opisującej a wartościami współczynników korelacji:_**
- Jeśli dane są wynikiem równomiernego spróbkowania funkcji o punkcie przegięcia w środku przedziału próbkowania, wartość współczynnika korelacji Spearmana jest co do wartości bezwzględnej mniejsza od wartości współczynnika korelacji Pearsona

- Jeśli dane są wynikiem równomiernego spróbkowania funkcji o niezerowych wartościach pierwszej i drugiej pochodnej, wartość współczynnika korelacji Kendalla jest równa wartości współczynnika korelacji Pearsona
